```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
    # Initialize min and max with the first element
    min_val = arr[0]
    max_val = arr[0]

    # Traverse the array
    for x in arr:
        if x < min_val:
            min_val = x
        if x > max_val:
            max_val = x

    return min_val, max_val

# Sample function call
arr = [12, 3, 4, 15, 45, 78, 450]
print("Min and Max:", max_min(arr))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0

    for x in arr:
        if x % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count

# Sample function call
arr = [12, 3, 4, 15, 45, 78, 450]
print("Even and Odd counts:", Counting_EvenOdd(arr))
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2   # middle index

        if arr[mid] == target:
            return mid            # element found, return index
        elif arr[mid] < target:
            low = mid + 1         # search in right half
        else:
            high = mid - 1        # search in left half

    return -1   # element not found

```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]

arr = [1, 3, 5, 7, 9, 15, 16, 14, 45]

# Sort the array first
arr.sort()
print("Sorted array:", arr)

# Perform binary searches
print("Search 7:", Binary_Search(arr, 7))
print("Search 14:", Binary_Search(arr, 14))
print("Search 45:", Binary_Search(arr, 45))
print("Search 20:", Binary_Search(arr, 20))

```
